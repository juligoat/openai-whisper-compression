## Horizontal join:

```python
from PIL import Image, ImageDraw, ImageFont

# Open the two images
img1 = Image.open("l1_pruning_librispeech_clean.png")
img2 = Image.open("l1_pruning_librispeech_other.png")

# Define the padding (space) between images in pixels
padding = 50  # Adjust this value for desired spacing

# Define header parameters with larger values
header_height = 100  # Increased height for the header area
header_text = "L1 Unstructured Pruning Results on LibriSpeech Datasets"
font_size = 56  # Increased font size for more prominence

# Try multiple bold font options, as font file names vary by system
bold_font_options = [
    "Arial Bold.ttf",
    "Arial-Bold.ttf",
    "arialbd.ttf",  # Windows common name
    "Arial-BoldMT.ttf",  # Mac common name
    "DejaVuSans-Bold.ttf",  # Common on Linux
    "FreeSansBold.ttf",  # Another Linux option
]

# Try to load a bold font
font = None
for font_name in bold_font_options:
    try:
        font = ImageFont.truetype(font_name, font_size)
        print(f"Successfully loaded bold font: {font_name}")
        break  # Exit the loop if a font is successfully loaded
    except OSError:
        continue  # Try the next font option

# If no bold font was found, fall back to default
if font is None:
    print("No bold fonts found, using default font")
    font = ImageFont.load_default()

# Calculate dimensions for the new combined image
total_width = img1.width + padding + img2.width
total_height = max(img1.height, img2.height) + header_height

# Create a new empty image (white background)
combined_img = Image.new("RGB", (total_width, total_height), color="white")

# Create a drawing object
draw = ImageDraw.Draw(combined_img)

# Calculate text position to center it
text_width = draw.textlength(header_text, font=font)
text_x = (total_width - text_width) // 2
text_y = (header_height - font_size) // 2  # Center vertically in header area

# Draw the header text
draw.text((text_x, text_y), header_text, fill="black", font=font)

# Paste the images below the header
img1_y = header_height
img2_y = header_height

combined_img.paste(img1, (0, img1_y))
combined_img.paste(img2, (img1.width + padding, img2_y))

# Save the result
combined_img.save("l1_pruning_combined.png")
```

## Vertical Join:

```python
from PIL import Image

# Open the two images
img1 = Image.open("l1_pruning_librispeech_clean_detailed.png")
img2 = Image.open("l1_pruning_librispeech_other_detailed.png")

# Define the padding (space) between images in pixels
# Increased padding for more space between the images
padding = 80  # Vertical space between images

# Calculate dimensions for the new combined image
# For vertical stacking, width is the max width of the two images
# Height is the sum of both images' heights plus padding between them
max_width = max(img1.width, img2.width)
total_height = img1.height + padding + img2.height

# Create a new empty image (white background)
combined_img = Image.new("RGB", (max_width, total_height), color="white")

# Calculate positions for pasting images
# For vertical stacking, we center each image horizontally if they have different widths
img1_x = (max_width - img1.width) // 2
img1_y = 0  # Start the first image at the top

img2_x = (max_width - img2.width) // 2
img2_y = img1.height + padding  # Position second image below first with padding

# Paste the images
combined_img.paste(img1, (img1_x, img1_y))
combined_img.paste(img2, (img2_x, img2_y))

# Save the result
combined_img.save("l1_pruning_combined_vertical.png")
print("Combined vertical image created successfully as 'l1_pruning_combined_vertical.png'")
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd
from matplotlib import gridspec

# Create the dataframes for both datasets with decoder-only pruning values
clean_data = {
    "Pruning_Percentage": [0, 10, 20, 30, 40, 50],
    "WER": [
        3.445414600109378,
        3.402040469949271,
        3.420898787410187,
        3.858411752503442,
        15.105512286193825,
        100.0,
    ],
}
clean_df = pd.DataFrame(clean_data)

other_data = {
    "Pruning_Percentage": [0, 10, 20, 30, 40, 50],
    "WER": [
        11.637230059377481,
        11.8887334064521,
        12.25558791271132,
        12.981732914791422,
        30.106652547180513,
        100.0,
    ],
}
other_df = pd.DataFrame(other_data)


# Function to create consistent charts
def create_chart(df, title, ylim_main, ylim_detail, save_filename):
    # Set up figure with two subplots
    fig = plt.figure(figsize=(12, 8), dpi=300)
    gs = gridspec.GridSpec(2, 1, height_ratios=[2, 1], hspace=0.4)

    # Main plot for full range
    ax0 = plt.subplot(gs[0])
    ax0.plot(
        df["Pruning_Percentage"],
        df["WER"],
        "o-",
        color="#2471A3",
        linewidth=2.5,
        markersize=8,
        markerfacecolor="white",
        markeredgecolor="#2471A3",
        markeredgewidth=2,
    )

    # Add WER values on top of each data point in the main plot
    for i, (x, y) in enumerate(zip(df["Pruning_Percentage"], df["WER"])):
        ax0.annotate(
            f"{y:.2f}%",
            xy=(x, y),
            xytext=(0, 7),  # 7 points vertical offset
            textcoords="offset points",
            ha="center",
            fontsize=9,
            fontweight="bold",
        )

    # Set up the main plot
    ax0.set_ylabel("Word Error Rate (%)", fontsize=12, fontweight="bold", labelpad=10)
    ax0.set_xlim(-5, 55)
    ax0.set_ylim(*ylim_main)
    ax0.set_title(title, fontsize=16, fontweight="bold", pad=15)
    ax0.set_xticks(df["Pruning_Percentage"])
    ax0.xaxis.set_major_formatter(mtick.PercentFormatter())
    ax0.grid(True, linestyle="--", alpha=0.3)

    # Show critical threshold with vertical line and annotation
    ax0.axvspan(30, 40, alpha=0.15, color="red")
    y_arrow = (ylim_main[1] - ylim_main[0]) * 0.65 + ylim_main[0]
    ax0.annotate(
        "Critical threshold",
        xy=(35, y_arrow * 0.8),
        xytext=(25, y_arrow),
        arrowprops=dict(arrowstyle="->", color="#C0392B", lw=1.5),
        fontsize=11,
        color="#C0392B",
        fontweight="bold",
    )

    # Detail view for the 0-30% range (before critical degradation)
    ax1 = plt.subplot(gs[1])
    ax1.plot(
        df["Pruning_Percentage"][:4],
        df["WER"][:4],
        "o-",
        color="#2471A3",
        linewidth=2.5,
        markersize=8,
        markerfacecolor="white",
        markeredgecolor="#2471A3",
        markeredgewidth=2,
    )

    # Add WER values on top of each data point in the detail view
    for i, (x, y) in enumerate(zip(df["Pruning_Percentage"][:4], df["WER"][:4])):
        ax1.annotate(
            f"{y:.2f}%",
            xy=(x, y),
            xytext=(0, 7),  # 7 points vertical offset
            textcoords="offset points",
            ha="center",
            fontsize=9,
            fontweight="bold",
        )

    # Set up the detail plot
    ax1.set_xlabel("Pruning Percentage", fontsize=12, fontweight="bold", labelpad=10)
    ax1.set_ylabel("WER (%)", fontsize=12, fontweight="bold", labelpad=10)
    ax1.set_title("Detail View (0-30% Pruning)", fontsize=12, fontweight="bold")
    ax1.set_xlim(-5, 35)
    ax1.set_ylim(*ylim_detail)
    ax1.set_xticks(df["Pruning_Percentage"][:4])
    ax1.xaxis.set_major_formatter(mtick.PercentFormatter())
    ax1.grid(True, linestyle="--", alpha=0.3)

    plt.tight_layout(pad=2.0)
    plt.savefig(save_filename, bbox_inches="tight", dpi=300)
    plt.close()


# Create both charts with consistent dimensions
create_chart(
    clean_df,
    "Decoder-only Pruning: Librispeech Clean",
    [0, 110],
    [3, 5],
    "l1_pruning_librispeech_clean_decoder.png",
)
create_chart(
    other_df, "Librispeech Other", [0, 110], [10, 15], "l1_pruning_librispeech_other_decoder.png"
)

# Now combine the consistently sized images
from PIL import Image

# Open the two images
img1 = Image.open("l1_pruning_librispeech_clean_decoder.png")
img2 = Image.open("l1_pruning_librispeech_other_decoder.png")

# Define the padding between images in pixels
padding = 85

# Calculate dimensions for the new combined image
max_width = max(img1.width, img2.width)
total_height = img1.height + padding + img2.height

# Create a new empty image (white background)
combined_img = Image.new("RGB", (max_width, total_height), color="white")

# Calculate positions for pasting images
img1_x = (max_width - img1.width) // 2
img1_y = 0

img2_x = (max_width - img2.width) // 2
img2_y = img1.height + padding

# Paste the images
combined_img.paste(img1, (img1_x, img1_y))
combined_img.paste(img2, (img2_x, img2_y))

# Save the result
combined_img.save("l1_pruning_decoder_combined_vertical.png")
print("Combined vertical image created successfully as 'l1_pruning_decoder_combined_vertical.png'")
```
